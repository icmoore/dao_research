### Analyze OHM Bond Polices
* Data: https://dune.xyz/shadow/Olympus-Policy

```python
import numpy as np
import pandas as pd
import datetime
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
def formatDates(arr,date_format="%Y-%m-%d %H:%M"):
    return [datetime.datetime.strptime(d, date_format) for d in arr]

def stripCharacter(arr,c,r=''):
    return [e.replace(c, r) for e in arr]

def str2Float(arr):
    return [float(e) for e in arr]

def getDate(t, strt_date):
    return datetime.datetime.strptime(strt_date, "%Y-%m-%d") + datetime.timedelta(days=t) 
```

```python
dai_bond = pd.read_csv('../resources/dai_bond.csv')
dai_bond = dai_bond.assign(evt_block_time = stripCharacter(dai_bond['evt_block_time'].values,'/','-'))
dai_bond = dai_bond.assign(evt_block_time = formatDates(dai_bond['evt_block_time'].values))
dai_bond = dai_bond.assign(initialBCV = str2Float(dai_bond['initialBCV'].values))
dai_bond = dai_bond.assign(newBCV = str2Float(dai_bond['newBCV'].values))
dai_bond = dai_bond.sort_values(by='evt_block_time')
dai_bond.head(5)
```

```python
frax_bond = pd.read_csv('../resources/frax_bond.csv')
frax_bond = frax_bond.assign(evt_block_time = stripCharacter(frax_bond['evt_block_time'].values,'/','-'))
frax_bond = frax_bond.assign(evt_block_time = formatDates(frax_bond['evt_block_time'].values))
frax_bond = frax_bond.assign(initialBCV = str2Float(frax_bond['initialBCV'].values))
frax_bond = frax_bond.assign(newBCV = str2Float(frax_bond['newBCV'].values))
frax_bond = frax_bond.sort_values(by='evt_block_time')
frax_bond.head(5)
```

```python
ohm_frax_bond = pd.read_csv('../resources/ohm_frax_bond.csv')
ohm_frax_bond = ohm_frax_bond.assign(evt_block_time = stripCharacter(ohm_frax_bond['evt_block_time'].values,'/','-'))
ohm_frax_bond = ohm_frax_bond.assign(evt_block_time = formatDates(ohm_frax_bond['evt_block_time'].values))
ohm_frax_bond = ohm_frax_bond.assign(initialBCV = str2Float(ohm_frax_bond['initialBCV'].values))
ohm_frax_bond = ohm_frax_bond.assign(newBCV = str2Float(ohm_frax_bond['newBCV'].values))
ohm_frax_bond = ohm_frax_bond.sort_values(by='evt_block_time')
ohm_frax_bond.head(5)
```

```python
ohm_dai_bond = pd.read_csv('../resources/ohm_dai_bond.csv')
ohm_dai_bond = ohm_dai_bond.assign(evt_block_time = stripCharacter(ohm_dai_bond['evt_block_time'].values,'/','-'))
ohm_dai_bond = ohm_dai_bond.assign(evt_block_time = formatDates(ohm_dai_bond['evt_block_time'].values))
ohm_dai_bond = ohm_dai_bond.assign(initialBCV = str2Float(ohm_dai_bond['initialBCV'].values))
ohm_dai_bond = ohm_dai_bond.assign(newBCV = str2Float(ohm_dai_bond['newBCV'].values))
ohm_dai_bond = ohm_dai_bond.sort_values(by='evt_block_time')
ohm_dai_bond.head(5)
```

```python
dates = dai_bond['evt_block_time'].values
initialBCV = dai_bond['initialBCV'].values
newBCV = dai_bond['newBCV'].values

fig = plt.figure(figsize=(12,5))
ax = fig.gca()
plt.plot(dates, initialBCV, color='darkorange', label="initialBCV")
plt.plot(dates, newBCV, color='blue', label="newBCV")

plt.title("DAI Bond Policy", fontsize=20)
plt.xlabel("Date", fontsize=14)
plt.ylabel("BCV", fontsize=14)
plt.legend();
fig.savefig('../img/dai_bond.png')
```
